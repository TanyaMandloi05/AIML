```python
import json
```

```python
def load_data(filename):
    with open(filename) as f:
        data = json.load(f)
        return data
```

```python
data = load_data("store.json")
print(data)
```

```python
# cleaning or structuring the data 
def clean_data(data):
    text_to_num = {"one":1, "two":2, "three":3, "four":4, "five":5}
    cleaned_data = []
    unique_user = set()
    for user in data:
        #clean rating 
        raw_rating = str(user["rating"]).strip().lower()
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        print(raw_rating)
        user["rating"] = raw_rating

        # cleaning age
        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = None
        if(user["name"].strip() in unique_user):
            continue
        unique_user.add(user["name"])
        cleaned_data.append(user)
        
    return cleaned_data
```

```python
cleaned_data = clean_data(data)
print(cleaned_data)
```

```python
# get meaningful insights from data
def get_insights(data):
    # average rating
    total_rating = 0
    for user in data:
        total_rating += float(user["rating"])

    print(f"average rating comes out to be {total_rating/len(data)}")

    # percentage of poor rating
    poor_rating = 0
    for user in data:
        if(float(user["rating"]) < 3):
            poor_rating += 1
    print(f"user with poor rating {poor_rating/len(data) * 100} % ")
```

```python
get_insights(cleaned_data)
```

```python
# recommendation system
def get_recommendation(data):
    recommendation = []
    for user in data:
        curr_user = {}
        curr_user["name"] = user["name"]
        if(float(user["rating"]) >= 4):
            curr_user["brand"] = "Apple"
        else:
            curr_user["brand"] = "samsung"
        recommendation.append(curr_user)
    return recommendation
```

```python
get_recommendation(cleaned_data)
```
