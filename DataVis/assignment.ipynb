```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(loc=70, scale=10, size=200)
mean = data.mean()
plt.hist(data, edgecolor="black", bins=15)
plt.axvline(x=mean, color="red", linewidth=3, label="mean value")
plt.xlabel("scores")
plt.ylabel("Frequency of scores")
plt.title("histogram of scores with mean line")
plt.legend()
plt.show()
```

```python
# Q3. Seaborn Dataset – Tips (Boxplot)
# Using the tips dataset:
# Create a boxplot of total_bill vs day.
# Add hue = "sex".
# Identify which day has the highest median bill.
# Comment on the presence of outliers.

sns.get_dataset_names()
tips = sns.load_dataset("tips")
sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex"
)
```

```python
# Q4. Seaborn Dataset – Tips (Scatter Plot)
# Using the tips dataset:
# Create a scatter plot of total_bill vs tip.
# Color the points by sex.
# Use different markers for smoker.
# Add a title and axis labels.
# Analyze: Do smokers tend to tip differently?

tips = sns.load_dataset("tips")
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    style="smoker"
)

plt.title("total bill vs tips")
plt.xlabel("total bills of all customer")
plt.ylabel("tips given by customer")
```

```python
# Q5. Seaborn Dataset – Tips (Pivot Table & Heatmap)
# Using the tips dataset:
# Create a pivot table of average tip:
# Rows: day
# Columns: time
# Plot the pivot table as a heatmap.
# Annotate the values on the heatmap.
# Interpret which day–time combination performs best.

tips = sns.load_dataset("tips")
tips_pivot = tips.pivot_table(index="day", columns="time", values="tip", aggfunc="mean",observed=False)
sns.heatmap(
    data=tips_pivot,
    annot=True,
    cmap="YlGn",
    cbar_kws={'label': 'Average Tip'}
)
plt.show()
```
