```python
# Scatter Plots

people = ["Person A", "Person B", "Person C", "Person D", "Person E",
          "Person F", "Person G", "Person H", "Person I", "Person J"]

age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]

blood_pressure = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

plt.title("Age vs BP")
# plt.scatter(age, blood_pressure, color="red", marker=">")
# colors=["red", "yellow", "blue", "purple", "orange","red", "yellow", "blue", "purple", "orange"]
# plt.scatter(age, blood_pressure, color=colors, marker=">")
# colors=["red" if x > 130 else "green" for x in blood_pressure ]
# plt.scatter(age, blood_pressure, color=colors, marker=">", s=blood_pressure, alpha=0.5)
colors=["red" if x > 130 else "green" for x in blood_pressure ]
plt.scatter(age, blood_pressure, cmap="YlOrRd", s=blood_pressure, c=blood_pressure)
plt.xlabel("age")
plt.ylabel("blood pressure")
plt.grid()
plt.colorbar(label="bp")

# annotations
for i in range(len(people)):
    plt.annotate(people[i], xy=(age[i], blood_pressure[i]), xytext=(age[i]+3, blood_pressure[i]))

plt.xlim(min(age), max(age)+12)
plt.ylim(min(blood_pressure), max(blood_pressure)+2)
```

```python
# Pie Charts
import matplotlib.pyplot as plt
expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]
plt.style.use("dark_background")
plt.figure(figsize=(5,5))
explode = [0,0,0,0.2,0]
plt.pie(amounts, labels=expenses, autopct="%1.1f%%", wedgeprops = {
    "edgecolor":"white",
    "linewidth": 2,
    "linestyle":"--"
    }, explode=explode,
       startangle=360,
       shadow=True)
plt.show()
```

```python
# subplots
import matplotlib.pyplot as plt
import numpy as np
x = [1,2,3,4,5,6,7,8]
y1 = [np.sqrt(i) for i in x] #square root 
y2 = [i*2  for i in x] #double value
y3 = [i ** 2 for i in x] # squares of numbers
y4 = [i ** 3 for i in x] #cube of number

plt.subplot(2,2,1)
plt.plot(x, y1)
plt.title("plot 1 - square root")

plt.subplot(2,2,2)
plt.plot(x, y2)
plt.title("plot 2 - double value")

plt.subplot(2,2,3)
plt.plot(x, y3)
plt.title("plot 3 - squares of number")

plt.subplot(2,2,4)
plt.plot(x, y4)
plt.title("plot 4 - cube of number")

plt.tight_layout()
```

```python
fig, ax = plt.subplots(2,2)
ax[0][0].plot(x, y1)
ax[0][0].set_title("plot 1 - square root")
fig.suptitle("Multiple plots")

ax[0][1].plot(x, y1)
ax[0][1].set_title("plot 1 - square root")

ax[1][0].plot(x, y1)
ax[1][0].set_title("plot 1 - square root")

ax[1][1].plot(x, y1)
ax[1][1].set_title("plot 1 - square root")

fig.tight_layout()
```

```python
# Task
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New York", "London", "Delhi", "Tokyo"]

temperatures = [
    [22, 23, 21, 24, 25],   # New York
    [18, 19, 17, 20, 21],   # London
    [30, 32, 31, 33, 34],   # Delhi
    [25, 26, 24, 27, 28]    # Tokyo
]

fig, ax = plt.subplots(2,2)
m = len(temperatures)
n = len(temperatures[0])
for i in range(m):
        row = i // 2
        col = i % 2
        ax[row][col].bar(days, temperatures[i])
        ax[row][col].set_title(cities[i])
        ax[row][col].set_xlabel("days")
        ax[row][col].set_ylabel("temperature")

fig.tight_layout()
```

```python
# Task
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New York", "London", "Delhi", "Tokyo"]

temperatures = [
    [22, 23, 21, 24, 25],   # New York
    [18, 19, 17, 20, 21],   # London
    [30, 32, 31, 33, 34],   # Delhi
    [25, 26, 24, 27, 28]    # Tokyo
]

fig, ax = plt.subplots(2,2)
ax = ax.flatten()
for i, ax in enumerate(ax):
    ax.plot(days, temperatures[i])
    ax.set_title(cities[i])

fig.suptitle("temperature of each cities")
fig.supxlabel("days")
fig.supylabel("temperature")
fig.tight_layout()
```
