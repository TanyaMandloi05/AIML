```python
import matplotlib.pyplot as plt
# plt is a module which provides various tools to visualize our data
```

```python
X = [1,2,3,4]
Y = [5,6,7,8]
plt.plot(X, Y)
plt.grid()
plt.show()
```

```python
oscar_movies = [
    "The Dark knight",
    "The hurt locker",
    "the kings speech",
    "the artist",
    "Argo"
]

year = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]
plt.plot(year, oscar_revenue)
plt.title("Oscar movie revenue in each year")
plt.xlabel("years")
plt.ylabel("revenue (in M $)")
```

```python
non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M
plt.plot(year, oscar_revenue, color="#CBC3E3", marker="<", linestyle="--", linewidth="3", label="oscar movie")
plt.plot(year, non_oscar_revenue, "^-y",label="non oscar movie")
plt.legend()
plt.xlabel("years")
plt.ylabel("revenue (in M $)")
plt.grid()
```

```python
with plt.xkcd():
    plt.plot(year, oscar_revenue, label="oscar movie")
    plt.plot(year, non_oscar_revenue, label="non oscar movie")
    plt.legend()
    plt.xlabel("years")
    plt.ylabel("revenue (in M $)")
    print(plt.style.available)
    plt.style.use('default')
    
    # plt.style.use('dark_background')
    # plt.style.use('ggplot')/
    # plt.grid(color="black", linestyle="--")
    plt.grid()
    plt.show()
```
