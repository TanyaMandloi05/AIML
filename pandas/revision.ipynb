```python
# group by and aggregate function
raw_data = pd.read_csv("raw_data.csv")
# we use group by function to split data into multiple groups so that we can apply some aggregation functions
# for example if we want mean salary based on country 
raw_data.groupby("country")["income"].mean()

# similarily based on male and female 
raw_data.groupby("gender")["income"].mean()

# if we want to apply multiple aggregation like min, max, count, sum together we can use agg
raw_data.groupby("country")["income"].agg(["mean", "max", "min"])

# we can also rename them 
raw_data.groupby("country")["income"].agg(mean_salary="mean", max_salary="max", min_salary="min")

# aggregation on multiple columns 
raw_data.groupby("country").agg(
    avg_income=("income","mean"), 
    ave_age=("age","mean")
)
```

```python
# reshaping methods 
# melt convert wide data into long data
import pandas as pd;
df = pd.DataFrame({
    "country":["USA", "India", "Canada", "China"],
    "year":[2021, 2022, 2026, 2025],
    "sales":[100, 120, 90, 110],
    "profit":[22, 23, 24, 25],
})

melted_df = df.melt(
    id_vars=["country","year"],
    value_vars=["sales", "profit"],
    var_name="metrics",
    value_name="value"
)

melted_df

# pivot - convert long format into wide format
original_data = melted_df.pivot(
    index = ["country", "year"],
    columns = "metrics",
    values = "value"
)

original_data
```

```python
# combining and joining methods
df_customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4],
    "name": ["Adam", "Bob", "Charlie", "Dave"]
})

df_orders = pd.DataFrame({
    "order_id": [101, 102, 103, 104],
    "customer_id": [2, 1, 4, 5],
    "amount": [250, 120, 300, 180]
})

pd.merge(df_customers, df_orders, on="customer_id") #innner joins (join only common values
pd.merge(df_customers, df_orders, on="customer_id" , how="left") #keep common plus whole left data
pd.merge(df_customers, df_orders, on="customer_id", how="right") #keep common plus right data
pd.merge(df_customers, df_orders, on="customer_id", how="outer") #keep both data whole
```
